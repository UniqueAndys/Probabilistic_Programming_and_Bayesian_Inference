## 1. Poisson分布 
$Z \sim Poi(\lambda)$  
概率质量方程 $P(Z=k)=\frac{\lambda^k e^{-\lambda}}{k!},k=0,1,2,\dots$  
$E[Z|\lambda]=\lambda$

```python
figsize(12.5, 4)
import scipy.stats as stats
a = np.arange(16)
poi = stats.poisson
lambda_ = [1.5, 4.25]
colors = ["#348ABD", "#A60628"]

# alpha 颜色深浅
# lw line width
plt.bar(a, poi.pmf(a, lambda_[0]), color=colors[0],
        label="$\lambda = %.1f$" % lambda_[0], alpha=0.60,
        edgecolor=colors[0], lw="3")

plt.bar(a, poi.pmf(a, lambda_[1]), color=colors[1],
        label="$\lambda = %.1f$" % lambda_[1], alpha=0.60,
        edgecolor=colors[1], lw="3")

plt.xticks(a + 0.4, a)
plt.legend()
plt.ylabel("probability of $k$")
plt.xlabel("$k$")
plt.title("Probability mass function of a Poisson random"+
          "variable; differing $\lambda$ values");
```

## 2. 指数分布  
$Z \sim Exp(\lambda)$  
概率密度函数 $f_z(z|\lambda) = \lambda e^{-\lambda z}, z\ge 0$  
$E[Z|\lambda] = 1 / \lambda$

```python
a = np.linspace(0, 4, 100)
expo = stats.expon
lambda_ = [0.5, 1]

for l, c in zip(lambda_, colors):
    plt.plot(a, expo.pdf(a, scale=1./l), lw=3,
            color=c, label="$\lambda = %.1f$"%l)
    plt.fill_between(a, expo.pdf(a, scale=1./l),
                    color=c, alpha=.33)
    
plt.legend()
plt.ylabel("Probability density function at $z$")
plt.xlabel("$z$")
plt.ylim(0, 1.2)
plt.title("Probability density function of an Exponential"+
         " random variable, differing $\lambda$ values");
```

## 实例：从短信数据推断行为

```python
figsize(12.5, 3.5)
count_data = np.loadtxt("./data/txtdata.csv")
n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data, color="#348ABD")
plt.xlabel("Time (days)")
plt.ylabel("count of text-msgs received")
plt.title("Did the user's texting habits change over time?")
plt.xlim(0, n_count_data)
```

$C_i$ 表示第 $i$ 天的短信条数：$C_i \sim Poisson(\lambda)$  
$$
\lambda =
\begin{cases}
\lambda_1 & \text{if } t \lt \tau \cr
\lambda_2 & \text{if } t \ge \tau
\end{cases}
$$
$\lambda_1 \sim Exp(\alpha)$  
$\lambda_2 \sim Exp(\alpha)$  
$\alpha$ 被称为超参数或者父变量， $\frac{1}{N}\sum^N_{i=0}C_i \approx E[\lambda|\alpha]=\frac{1}{\alpha}$  
$\tau \sim DiscreteUniform(1, 70) \Rightarrow P(\tau=k) = 1 / 70$

```python
import pymc as pm
print("count data mean:", count_data.mean())
alpha = 1.0 / count_data.mean()
# 产生对应于参数lambda_1, lambda_2, tau的PyMC随机变量
lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)
tau = pm.DiscreteUniform("tau", lower=0, upper=n_count_data)

#print("Random output:", tau.random(), tau.random(), tau.random())
#print("Random output:", lambda_1.random(), lambda_1.random(), lambda_1.random())
#print("Random output:", lambda_2.random(), lambda_2.random(), lambda_2.random())
```

```python
# @pm.deterministic是一个标识符，告诉PyMC这是一个确定性函数，
# 即如果函数的输入为确定的话，那函数的结果也是确定的。
@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    # 对应于之前的lambda函数
    out = np.zeros(n_count_data)
    out[:tau] = lambda_1
    out[tau:] = lambda_2
    return out

# lambda_ 数据产生机制
# value=count_data 数据
# observed=True stay fix in our analysis
observation = pm.Poisson("obs", lambda_, 
                         value=count_data, 
                         observed=True)
# 收集所有的变量，创建一个Model实例
model = pm.Model([observation, lambda_1, lambda_2, tau])
```
